```python
name_list = ["x1", "x2", "x3", "x4", "x5", "x6", "x7", "x8", "x9", "y"]
```

```python
df_train = pd.read_csv('./Breast cancer/Breast-cancer-wisconsin.data', names = name_list)
df_train.head()
```

```python
df_train.y.value_counts()
```

```python
df_train.x6 = df_train.x6.replace("?","1")
```

```python
df_train.x6 = df_train.x6.astype(np.int64)
```

```python
from sklearn import preprocessing

# for colname in df_train.columns:
#     encoder = preprocessing.LabelEncoder()
#     df_enc = encoder.fit_transform(df_train[colname])
#     df_train[colname] = df_enc
encoder = preprocessing.LabelEncoder()
df_enc = encoder.fit_transform(df_train["y"])
df_train["y"] = df_enc
```

```python
df_train.head()
```

```python
lab = df_train.y
del df_train["y"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_train.values, lab.values, test_size=0.75, random_state=42)
```

```python
y_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae


train_acc_list = np.zeros((50,1))
test_acc_list = np.zeros((50,1))
for iter in range(1000):
    for m in range(1,50):
        print('iter: ' + str(iter) + ', m: ' + str(m))

        inds = np.random.choice(range(X_train.shape[0]), m, replace=False)
        xtrain = X_train[inds]
        ytrain = y_train[inds]

        clf = GaussianNB(priors=None)
        clf.fit(xtrain, np.ravel(ytrain))

        ypred_train = clf.predict(xtrain)
        train_acc_list[m] = train_acc_list[m] + mse(ytrain, ypred_train)

        ypred = clf.predict(X_test)
        test_acc_list[m] = test_acc_list[m] + mse(y_test,ypred)
        
train_acc_list = train_acc_list/1000.0
test_acc_list = test_acc_list/1000.0
```

```python
from sklearn.linear_model import LogisticRegression

train_acc_list_lr = np.zeros((50,1))
test_acc_list_lr = np.zeros((50,1))
for iter in range(1000):
    for m in range(2,50):
        print('iter: ' + str(iter) + ', m: ' + str(m))

        inds = np.random.choice(range(X_train.shape[0]), m, replace=False)
        xtrain = X_train[inds]
        ytrain = y_train[inds]
        z = np.ravel(ytrain)
        l = len(set(z))
        while(l < 2):
            inds = np.random.choice(range(X_train.shape[0]), m, replace=False)
            xtrain = X_train[inds]
            ytrain = y_train[inds]
            z = np.ravel(ytrain)
            l = len(set(z))

        
        clf = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial').fit(xtrain, np.ravel(ytrain))

        ypred_train = clf.predict(xtrain)
        train_acc_list_lr[m] = train_acc_list_lr[m] + mse(ytrain, ypred_train)

        ypred = clf.predict(X_test)
        test_acc_list_lr[m] = test_acc_list_lr[m] + mse(y_test,ypred)
        
train_acc_list_lr = train_acc_list_lr/1000.0
test_acc_list_lr = test_acc_list_lr/1000.0
```
